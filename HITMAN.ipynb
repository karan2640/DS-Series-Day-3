```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.countplot(y='Against', data=df, order=df['Against'].value_counts().index, palette='crest')
plt.title('Rohit Sharma - Centuries by Opponent')
plt.xlabel('Number of Centuries')
plt.ylabel('Opponent')
plt.tight_layout()
plt.show()
```

```python
position_counts = df['Position'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(position_counts, labels=position_counts.index, autopct='%1.1f%%', startangle=140,
        colors=sns.color_palette('Set2'))
plt.title('Rohit Sharma Centuries by Batting Position')
plt.axis('equal')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year

centuries_per_year = df['Year'].value_counts().sort_index()

plt.figure(figsize=(10,5))
sns.lineplot(x=centuries_per_year.index, y=centuries_per_year.values, marker='o', color='orange')
plt.title('Rohit Sharma - Centuries Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Centuries')
plt.grid(True)
plt.show()
```

```python
top_grounds = df['Ground'].value_counts().head(5)

plt.figure(figsize=(8,5))
sns.barplot(x=top_grounds.values, y=top_grounds.index, palette='magma')
plt.title('Top Grounds Where Rohit Scored Most Centuries')
plt.xlabel('Number of Centuries')
plt.ylabel('Ground')
plt.show()
```

```python
results = df['Result'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(results, labels=results.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Match Result When Rohit Scored a Century')
plt.axis('equal')
plt.show()
```
